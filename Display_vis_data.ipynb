```python
from gensim import corpora
import cPickle as pickle
import pyLDAvis.gensim
import pyLDAvis

NUM_TOPICS=30
```

```python
# # if you want to build the vis_data from the LDA object
lda_filename = 'lda_ortho_'+str(NUM_TOPICS)+'t_pos.p'
print lda_filename
lda = pickle.load(open(lda_filename,'rb'))

corpus = corpora.MmCorpus('orthopedists_bow.mm')
dictionary = corpora.Dictionary.load('orthopedists.dict')
```

```python
lda.show_topics(0,5)#topic_id, topn
# Return a list of (words_probability, word) 2-tuples for the most probable words in topic topicid.
```

```python
# build vis_data from LDA
vis_data = pyLDAvis.gensim.prepare(lda, corpus, dictionary)
pyLDAvis.display(vis_data)#html object
```
